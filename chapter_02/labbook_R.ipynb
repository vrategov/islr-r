## Graphics

```r
# basic plot command
x = rnorm(100)
y = rnorm(100)
plot(x, y)
```

```r
# adding axis labels and title
plot(x,
     y, 
     xlab="this is the x-axis", 
     ylab="this is the y-axis",
     main="Plot of X vs Y")
```

```r
# save plot in a pdf file
pdf("Figure.pdf")
plot(x, y, col="green")
dev.off()
```

```r
# creating a sequence of numbers - using seq() and the shorthand :
x = seq(1, 10)
print(x)
x = 1:10
print(x)
x = seq(-pi, pi, length=50)
print(x)
```

```r
# making contour plots
y = x
f = outer(x, y, function(x, y)cos(y)/(1+x^2))  # calling a custom function in the same line of code 
contour(x, y, f)
contour(x, y, f, nlevels=45, add=T)  # nlevels corresponds to the level of granularity we want to operate, while add=T tells R to add to current plot
```

```r
fa = (f - t(f)) / 2  # here the function t() transpose the original matrix f
contour(x, y, f, nlevels=15)
```

```r
# in the contour map what we lack is colour - we can get with the help of image() and create a heatmap
image(x, y, fa)
```

```r
# a 3D plot could be done with persp()
persp(x, y, fa)
```

```r
persp(x, y, fa, theta=30)
```

```r
persp(x, y, fa, theta=30, phi=20)
```

```r
persp(x, y, fa, theta=30, phi=70)
```

```r
persp(x, y, fa, theta=30, phi=40)
```
